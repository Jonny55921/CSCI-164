```python
# prompt: Can you alter the size of this image to fit in a simple window for the display image function

import warnings
from transformers import pipeline
from transformers.utils import logging
from IPython.display import Image
from PIL import Image as PILImage
warnings.filterwarnings(action = 'ignore')
logging.set_verbosity(40)
# file 1
#https://drive.google.com/file/d/1VfRCDm6nZDUY6kHYPIbKlYRIBpxWHP2i/view?usp=drive_link
!gdown 1VfRCDm6nZDUY6kHYPIbKlYRIBpxWHP2i
file_name1 = "corgis.jpg"

# Resize the image
try:
  img = PILImage.open(file_name1)
  img = img.resize((500, 500)) # Example size, adjust as needed
  img.save("resized_corgis.jpg") # Save as a new file
  Image("resized_corgis.jpg") #Display the resized image
except FileNotFoundError:
  print(f"Error: Image file '{file_name1}' not found.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
